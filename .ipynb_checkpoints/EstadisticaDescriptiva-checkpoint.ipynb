```python
# Importar librerías a utilizar
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns
```

```python
# Leer archivo "World population.csv" y cargarlo a un dataframe
df = pd.read_csv("DataSets/AS3.9/World population.csv")
```

```python
# Ver las primeras 5 líneas del archivo
df.head()
```

```python
# Reemplazar espacios por NaN
df = df.replace(r'^\s*$', np.NaN, regex=True)
df.head()
```

```python
# Verificar cuantos datos perdidos hay
df.isna().sum()
```

```python
# Usar dropna para eliminar las filas donde hay datos vacíos
df = df.dropna(how = 'any')
df.head()
```

```python
# Ver cuantos paises quedan
index = df.index
number_of_rows = len(index)
print(number_of_rows)
```

```python
# Cambiar el tipo de las columnas a numerico
df = df.astype({
    'GDP': 'int64',
    'Live births number': 'int64',
    'Live births rate': 'float64',
    'Deaths number': 'int64',
    'Death rate': 'float64',
    'Rate of increase': 'float64',
    'Infant deaths number': 'int64',
    'Infant deaths rate (per 1000 births)': 'float64',
    'Total fertility rate': 'float64'
    })
df.dtypes
```

```python
# Encontrar rangos de las variables
print("Date: Min -", df['Date'].min(), ", Max -", df['Date'].max())
print("GDP: Min -", df['GDP'].min(), ", Max -", df['GDP'].max())
print("GNI: Min -", df['GNI'].min(), ", Max -", df['GNI'].max())
print("Live births number: Min -", df['Live births number'].min(), ", Max -", df['Live births number'].max())
print("Live births rate: Min -", df['Live births rate'].min(), ", Max -", df['Live births rate'].max())
print("Deaths number: Min -", df['Deaths number'].min(), ", Max -", df['Deaths number'].max())
print("Death rate: Min -", df['Death rate'].min(), ", Max -", df['Death rate'].max())
print("Rate of increase: Min -", df['Rate of increase'].min(), ", Max -", df['Rate of increase'].max())
print("Infant deaths number: Min -", df['Infant deaths number'].min(), ", Max -", df['Infant deaths number'].max())
print("Infant deaths rate (per 1000 births): Min -", df['Infant deaths rate (per 1000 births)'].min(), ", Max -", df['Infant deaths rate (per 1000 births)'].max())
print("Life expectancy at birth male: Min -", df['Life expectancy at birth male'].min(), ", Max -", df['Life expectancy at birth male'].max())
print("Life expectancy at birth female: Min -", df['Life expectancy at birth female'].min(), ", Max -", df['Life expectancy at birth female'].max())
print("Total fertility rate: Min -", df['Total fertility rate'].min(), ", Max -", df['Total fertility rate'].max())
```
