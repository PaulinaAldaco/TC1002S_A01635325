```python
# Importar librerías a utilizar
import numpy as np
import matplotlib.pyplot as plt 

import pandas as pd  
import seaborn as sns
```

```python
# Leer archivo "World population.csv" y cargarlo a un dataframe
df = pd.read_csv("DataSets/AS3.9/World population.csv")
```

```python
# Reemplazar espacios por NaN
df = df.replace(r'^\s*$', np.NaN, regex=True)
```

```python
# Usar dropna para eliminar las filas donde hay datos vacíos
df = df.dropna(how = 'any')
```

```python
# Cambiar el tipo de las columnas a numerico
df = df.astype({
    'GDP': 'int64',
    'Live births number': 'int64',
    'Live births rate': 'float64',
    'Deaths number': 'int64',
    'Death rate': 'float64',
    'Rate of increase': 'float64',
    'Infant deaths number': 'int64',
    'Infant deaths rate (per 1000 births)': 'float64',
    'Total fertility rate': 'float64'
    })
```

```python
# Para cada columna en df, crear un boxplot, excepto si la columna es Country o Continent
for column in df:
    if(column == 'Country' or column == 'Continent'):
        continue
    plt.figure()
    df.boxplot([column])
```

```python
# Para cada columna en df, crear un histograma, excepto si la columna es Country o Continent
for column in df:
    if(column == 'Country' or column == 'Continent'):
        continue
    sns.distplot(df[column])
    plt.show()
```

```python
# Obtener una matriz de correlación
correlation_matrix = df.corr().round(2)
# Usar la matriz de correlación para dibujar un mapa de calor
# annot = True para mostrar los valores dentro del cuadro
plt.subplots(figsize=(15,10)) # Hacer tamaño mas grande
sns.heatmap(data=correlation_matrix, annot=True)
```

```python
# Los boxplots y los histogramas muestran de forma mas visual en donde se concentran los datos, si hay o no 
# valores atipicos, y en donde se encuentran estos valores. Algo que es claro al analizar estas graficas es
# que hay un pais con un valor exagerado en comparacion a los demas paises en cuanto a muertes de infantes y
# tasa de mortalidad infantil. Lo mismo sucede en el campo de GDP. Estos valores extremos hacen dificl
# visualizar a los demas valores. De los histogramas se puede ver que los valores para los campos de GDP,
# numero de nacimientos, numero de muertes, numero de muertes infantiles, y tasa de muertes infantiles, se
# concentran alrededor de los mismos valores para la mayoria de los paises (exceptuando los valores atipicos).

# Analizando el heatmap, podemos identificar que variables presentan alguna correlacion. Las mas fuertes se dan
# entre la tasa de mortalidad infantil y la expectativa de vida de las mujeres, la tasa de mortalidad en general
# y la tasa de crecimiento, expectativa de vida de hombres y de mujeres, y el GDP y el numero de muertes. No
# se mencionan otros ejemplos de correlacion porque son obvios, por ejemplo, tasa de nacimientos y tasa de 
# fertilidad.

# Si se tuviera que eliminar variables, eliminaria las de numeros totales para dejar solo las tasas, ya que
# es un poco redundante tener ambos y las tasas pueden ser un mejor indicador de la situacion en el pais.
```
